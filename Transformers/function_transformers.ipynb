```python
import pandas as pd
import numpy as np

import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score

from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df= df[['age','avg_glucose_level','bmi','stroke']]
```

```python
df.head()
```

```python
X = df.drop('stroke',axis=1)
y = df['stroke']
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(df['age'])
plt.title('Age - density plot')

plt.subplot(122)
stats.probplot(df['age'], dist="norm", plot=plt)
plt.title('Age - QQ plot')

plt.show()
```

<h5>Let's create a function to display the above plots for all the features</h5>

```python
def plot_distribution(df, feature):
    
        plt.figure(figsize=(14,4))
        plt.subplot(121)
        sns.histplot(data=df,x=df[feature],kde=True)
        plt.title(feature+' - density plot')

        plt.subplot(122)
        stats.probplot(df[feature], dist="norm", plot=plt)
        plt.title(feature+' - QQ plot')
        
        plt.show()
```

```python
for feature in X.columns:
    plot_distribution(X,feature)
```

<h5> The histogram with density curve is visually representative of the how the feature values are distributed<.h5>

<h5> How to read the QQ plot? <br>
If the feature is normally distributed, the QQplot is a straight 45degree line.<br>
If it deviates upwards from the straight line (red), then the feature is right skewed (few very high values).
If it deviates downwards from the red line on the left side, then the feature is left skewed (few very low values)</h5>



<h4> Now, let's see the effect of transformers on the performace of the algorithms</h4>

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=2)
```

<h2>Before Transformation</h2>

```python
clf_lr = LogisticRegression()
clf_dt = DecisionTreeClassifier()
```

```python
clf_lr.fit(X_train, y_train)
clf_dt.fit(X_train, y_train)

y_pred_lr = clf_lr.predict(X_test)
y_pred_dt = clf_dt.predict(X_test)
```

```python
print("Logistic regression: {}".format(accuracy_score(y_test, y_pred_lr)))
print("Decision Tree: {}".format(accuracy_score(y_test, y_pred_dt)))
```

<h2>After Transformation</h2>

```python
trf = FunctionTransformer(func=np.log1p)

X_train_transform = trf.fit_transform(X_train)
X_test_transform = trf.transform(X_test)


clf_lr_2 = LogisticRegression()
clf_dt_2 = DecisionTreeClassifier()

clf_lr_2.fit(X_train_transform, y_train)
clf_dt_2.fit(X_train_transform, y_train)

y_pred_lr = clf_lr_2.predict(X_test_transform)
y_pred_dt = clf_dt_2.predict(X_test_transform)


print("Logistic regression: {}".format(accuracy_score(y_test, y_pred_lr)))
print("Decision Tree: {}".format(accuracy_score(y_test, y_pred_dt)))
```

```python

trf_3 = FunctionTransformer(lambda x:1/x)

X_train_transform_3 = trf_3.fit_transform(X_train)
X_test_transform_3 = trf_3.transform(X_test)


clf_lr_3 = LogisticRegression()
clf_dt_3 = DecisionTreeClassifier()

clf_lr_3.fit(X_train_transform_3, y_train)
clf_dt_3.fit(X_train_transform_3, y_train)

y_pred_lr_3 = clf_lr_3.predict(X_test_transform_3)
y_pred_dt_3 = clf_dt_3.predict(X_test_transform_3)


print("Logistic regression: {}".format(accuracy_score(y_test, y_pred_lr_3)))
print("Decision Tree: {}".format(accuracy_score(y_test, y_pred_dt_3)))
```

<h4> We see that applying transformation does not have any major effect on this dataset for both linear based model 
which is logistic regression and tree based model which is Decision tree</h4>
